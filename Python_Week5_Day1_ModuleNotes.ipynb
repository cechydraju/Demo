**Matplotlib:**

Matplotlib is  used to create 2D graphs and plots by using python scripts It provides a MATLAB-like interface and supports a wide variety of plots including line plots, scatter plots, bar plots, histograms, pie charts, and more. Matplotlib is highly customizable, allowing users to fine-tune every aspect of their plots. It is one of the most widely used libraries for data visualization in Python due to its versatility and ease of use.

**Seaborn:**

Seaborn is a statistical data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn simplifies the process of creating complex visualizations such as heatmaps, violin plots, box plots, and pair plots. It also offers built-in themes and color palettes to enhance the aesthetics of your plots. Seaborn is particularly useful for visualizing statistical relationships in datasets and is often used in conjunction with Pandas data structures.









**Key Differences between seaborn and matplotlib:**

**Defaults:** Seaborn applies default styling and aesthetics, while Matplotlib requires more manual configuration.

**Convenience:**  Seaborn often simplifies plot creation with less code.

**Types of Data:**

**1. Nominal Data**

**Definition:** Data that represents categories or labels without any inherent order. Values can't be ranked or compared mathematically.

**Examples:**

Gender (Male, Female, Other)

Eye color (Blue, Brown, Green)

Country of origin (India, USA, Japan)

Types of fruit (Apple, Banana, Orange)


**2. Ordinal Data**

**Definition:** Data that has a natural order or ranking. However, the differences between the values are not necessarily equal or meaningful.

**Examples:**

Survey ratings (Strongly Disagree, Disagree, Neutral, Agree, Strongly Agree)

Educational levels (High School, Bachelor's, Master's, PhD)

Performance ranking (First, Second, Third)

Clothing sizes (Small, Medium, Large)

**3. Discrete Data**

**Definition:** Data that represents counts of whole numbers. It can only take on specific values, with nothing in between.

**Examples:**

Number of children in a family

Number of cars on the road

Number of rooms in a house

Number of defects in a product

**4. Continuous Data**

**Definition:** Data that can take on any value within a range. It can be infinitely subdivided. Measurements belong to this category.

**Examples:**

Height (e.g., 5.8 feet)

Weight (e.g., 72 kilograms)

Temperature (e.g., 22.5 degrees Celsius)

Time (e.g., 3.45 seconds)

**Scatter Plot:**

A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship between two variables


```python
#Example of scatter plot to see the relationship between age and weight values
Age = [20, 24, 28, 32, 36]
Weight = [52, 59, 68, 72, 82]
plt.scatter(Age,Weight,marker = 'H',s=200,facecolor='Green',edgecolor='Blue')
plt.xlabel("Age in years")
plt.ylabel("Weight in Kg")
plt.title("Relationship between Age and Weight")
plt.show()
```

The great thing about the normally distributed values (or values in a bell-shaped curve) is that approximately

- $68 \text{%}$ of the values lie between $\mu - \sigma$ and $\mu + \sigma$. In other words, approx. $68 \text{%}$ of the values lie within **one-sigma** around the mean.

- $95 \text{%}$ of the values lie between $\mu - 2 \sigma$ and $\mu + 2 \sigma$. In other words, approx. $95 \text{%}$ of the values lie within **two-sigma** around the mean.

- $99.7 \text{%}$ of the values lie between $\mu - 3 \sigma$ and $\mu + 3 \sigma$. In other words, approx. $99.7 \text{%}$ of the values lie within **three-sigma** around the mean.

where $\mu$ and $\sigma$ are mean and standard deviation respectively.

```python
## example to show that 68% of the data lies in one -sigma
## creating the random values of data
import numpy as np
height_mean = 165
height_std = 15
a = np.random.normal(height_mean,height_std,size=10000)
a
```

```python
## checking if the data lies in 68% for one-sigma
import numpy as np
one_sigma=0
# a=tips["tip"]
for i in a:
  if i >=np.mean(a)-np.std(a) and i<=np.mean(a)+np.std(a):
    one_sigma+=1
one_sigma*100/len(a)
```

**Subplots:**

Subplots allow you to create multiple plots within a single figure

```python

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))  # Create figure and axes grid

# Plot 1: Distribution of total bill
sns.histplot(tips["total_bill"], ax=axes[0])
axes[0].set_title("Distribution of Total Bill")
axes[0].set_xlabel("ttt")

# Plot 2: Tip amount by sex
sns.boxplot(x="sex", y="tip", data=tips, ax=axes[1])
axes[1].set_title("Tip Amount by Sex")
plt.grid()  # Adjust spacing between subplots

# Plot 3: Tip amount by day
sns.violinplot(x="day", y="tip", data=tips, ax=axes[2])
axes[2].set_title("Tip Amount by Day")

plt.grid()  # Adjust spacing between subplots
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

sns.histplot(tips["total_bill"], ax=axes[0, 0])
sns.histplot(tips["tip"], ax=axes[0, 1])
sns.boxplot(x="sex", y="tip", data=tips, ax=axes[1, 0])
sns.violinplot(x="day", y="tip", data=tips, ax=axes[1, 1])

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

**Plotly:**

Plotly is a versatile library for creating interactive visualizations in Python. It supports a wide range of chart types including line charts, scatter plots, bar charts, histograms, 3D plots, and geographical maps. Plotly generates interactive plots that can be embedded in web applications or exported as standalone HTML files. It also offers APIs for other programming languages such as R, JavaScript, and Julia. Plotly is popular for its ability to create rich, interactive visualizations with minimal code and effort.

```python
import plotly.express as px

import seaborn as sns

# Load tips dataset from seaborn
tips = sns.load_dataset("tips")

# Scatter Plot
scatter_fig = px.scatter(tips, x='total_bill', y='tip', title='Scatter Plot of Total Bill vs Tip Amount')

scatter_fig.show()
```

```python
# Line Plot
line_fig = px.line(tips, x='total_bill', y='tip', title='Line Plot of Total Bill vs Tip Amount',labels={"total_bill":"total bill"})
line_fig.update_layout(height=400, width=800)
line_fig.show()
```

```python

# Bar Plot
barplot_fig = px.bar(tips.groupby('day')['total_bill'].mean().reset_index(), x='day', y='total_bill',
                     title='Average Total Bill by Day', labels={'total_bill': 'Average Total Bill ($)'}, color='day')

barplot_fig.show()
```

```python

# Pie Chart
piechart_fig = px.pie(tips, names='day', title='Distribution of Days', labels={'day': 'Day'})
piechart_fig.show()

```

```python
# Histogram
histogram_fig = px.histogram(tips, x='total_bill', title='Histogram of Total Bill Amounts', nbins=20)
histogram_fig.show()
```

```python

# Count Plot
countplot_fig = px.bar(tips['day'].value_counts().reset_index(), x='index', y='day',
                       title='Count of Observations by Day', labels={'index': 'Day', 'day': 'Count'})


countplot_fig.show()

```

```python

# Pair Plot
pairplot_fig = px.scatter_matrix(tips)
pairplot_fig.show()

```

```python
# Box Plot
boxplot_fig = px.box(tips, y='total_bill', title='Box Plot of Total Bill Amounts')
boxplot_fig.show()

```

```python

# Violin Plot
violinplot_fig = px.violin(tips, y='total_bill', title='Violin Plot of Total Bill Amounts')

violinplot_fig.show()

```

```python

# Heatmap
heatmap_data = tips.corr()
heatmap_fig = px.imshow(heatmap_data, title='Correlation Heatmap')

heatmap_fig.show()

```

**Note:** Plotly Express does not support arbitrary subplot capabilities

To create subplots using Plotly for the tips dataset, you can use the make_subplots function from Plotly's **plotly.subplots** module

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns

# Load tips dataset
tips = sns.load_dataset("tips")

# Create subplots
fig = make_subplots(rows=2, cols=2, subplot_titles=("Scatter Plot", "Bar Plot", "Histogram", "Box Plot"))

# Scatter Plot
scatter_trace = go.Scatter(x=tips["total_bill"], y=tips["tip"], mode="markers")
fig.add_trace(scatter_trace, row=1, col=1)
fig.update_xaxes(title_text="Total Bill ($)", row=1, col=1)
fig.update_yaxes(title_text="Tip ($)", row=1, col=1)

# Bar Plot
bar_trace = go.Bar(x=tips['day'].unique(), y=tips.groupby('day')['total_bill'].mean(), marker_color='skyblue')
fig.add_trace(bar_trace, row=1, col=2)
fig.update_xaxes(title_text="Day", row=1, col=2)
fig.update_yaxes(title_text="Average Total Bill ($)", row=1, col=2)

# Histogram
histogram_trace = go.Histogram(x=tips['total_bill'], marker_color='salmon')
fig.add_trace(histogram_trace, row=2, col=1)
fig.update_xaxes(title_text="Total Bill ($)", row=2, col=1)
fig.update_yaxes(title_text="Frequency", row=2, col=1)

# Box Plot
box_trace = go.Box(y=tips['total_bill'], boxmean=True, marker_color='lightgreen')
fig.add_trace(box_trace, row=2, col=2)
fig.update_xaxes(title_text="Total Bill ($)", row=2, col=2)
fig.update_yaxes(title_text="Total Bill ($)", row=2, col=2)

# Update layout
fig.update_layout(title="Subplots for Tips Dataset", height=600, width=800)

# Show plot
fig.show()
```
